```python
# Crea un nuevo conjunto de datos en donde elimina las filas vacías. Igualando todo el conjunto de datos a una misma cantidad de registros
datos.dropna()
```

```python
# Eliminamos los registros que están vacíos para tener una uniformidad de datos y que no hayan celdas vacías
datos_na = datos.dropna()
```

```python
datos_na.info()
```

```python
# Convertir la característica categórica (OCEAN PROXIMITY) a numérica
datos_na["ocean_proximity"]
```

```python
datos_na["ocean_proximity"].value_counts()
```

```python
# Dummies / One-Hot Enconding
"""
Lo que hace los Dummies o One-Hot Encondnig es agarrar esos datos

"""
# NEAR BAY INSALAND NEAR OCEAN
#     1        0        0
#     0        1        0
# Genera los dummies y los guarda en una variable
dummies = pd.get_dummies(datos_na["ocean_proximity"])
```

```python
# Añade los dummies al set de datos anterior
datos_na = datos_na.join(dummies)
```

```python
datos_na.head()
```

```python
# Eliminamos ocean_proximity
datos_na = datos_na.drop(["ocean_proximity"], axis=1)
```

```python
# Análisis, nuevas características
datos.corr()
```

```python
sb.set(rc={"figure.figsize": (15, 8)}) # Setea el tamaño del gráfico
# Crea un mapa de tipo "heat"
# annot=True coloca las anotaciones al lado del gráfico
# cmap="YlGnBu" setea los colores del gráfico
sb.heatmap(datos_na.corr() ,annot=True, cmap="YlGnBu")
```

```python
# Esto me permite saber cuan relacionados estan el resto de valores con respecto a median_house_value. Como se puede ver en los resultados:
# El valor medio de la casa esta relacionado con la media de ingresos (0.68) y la cantidad de habitaciones (0.13)
datos.corr()["median_house_value"].sort_values(ascending=False)
```

```python
# Crea un gráfico en donde el eje X corresponde al valor medio de las casas y el eje Y al ingresio medio de las personas
sb.scatterplot(x=datos_na["median_house_value"] ,y=datos_na["median_income"])
```

```python
# Creamos una nueva columna en el set de datos (datos_na) con una proporción  entre el total de habitaciones para dormir y el total de habitaciones
datos_na["bedroom_ratio"] = datos_na["total_bedrooms"] / datos_na["total_rooms"]
```

```python
sb.set(rc={"figure.figsize": (15, 8)})
sb.heatmap(datos_na.corr() ,annot=True, cmap="YlGnBu")
```

```python
# Separar las características de las etiquetas
X = datos_na.drop(["median_house_value"], axis=1) # Axis 1 = Eje X
Y = datos_na["median_house_value"]
```

```python
# Separar los datos en dos partes: Conjunto de entrenamiento y uno de pruebas
from sklearn.model_selection import train_test_split

X_entrenamiento, X_pruebas, Y_entrenamiento, Y_pruebas = train_test_split(X, Y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(X_entrenamiento, Y_entrenamiento)
```

```python
predicciones = modelo.predict(X_pruebas)
```

```python
comparativa = {"Predicción": predicciones, "Valor Real": Y_pruebas}
pd.DataFrame(comparativa)
```

```python
# Overfitting
print(modelo.score(X_entrenamiento, Y_entrenamiento))
print(modelo.score(X_pruebas, Y_pruebas))
```

```python
# Calcular el Error Medio
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(Y_pruebas, predicciones)
```

```python
mse
```

```python
# Calculamos la raíz cuadrada para tener valores más pequeños
rmse = np.sqrt(mse)
rmse
```
